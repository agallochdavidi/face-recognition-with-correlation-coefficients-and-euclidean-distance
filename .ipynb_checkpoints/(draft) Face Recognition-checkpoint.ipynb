```python
# Създаване на списък с имена на файловете в папката images
files = os.listdir('images')

# Зареждане на първото изображение и определяне на размера на матрицата за съхранение на сумата на изображенията
img = Image.open('images/'+files[0]).convert('L')
img_arr = np.array(img, dtype=float)
img_sum = np.zeros_like(img_arr, dtype=float)

# Добавяне на всички изображения към сумата и превръщането им в NumPy масиви
for file in files[1:]:
    img = Image.open('images/'+file).convert('L')
    img_arr = np.array(img, dtype=float)
    img_sum += img_arr

# Изчисляване на усредненото изображение
avg_img = img_sum / len(files)

# Прескалиране на изображението в диапазона [0, 255] и преобразуване в NumPy масив
avg_img = np.uint8(avg_img / np.max(avg_img) * 255)

# Плотване на усредненото изображение
plt.imshow(avg_img, cmap='gray')
plt.show()
```

Първоначално, изчисляваме средно квадратичната грешка (MSE) между средното изображение (avg_img) и входното изображение (image), което ни дава информация за колко далеч е входното изображение от средното.

След това, ако MSE е 0 (което би означавало, че двата файлове са абсолютно еднакви), то се прилага специално условие и се изчислява PSNR, като се дава стойност 100.

В противен случай, изчислява PSNR като се използва максималната стойност на пикселите (max_pixel) и се изчислява логаритъмичната стойност на коефициента на сигнал към шум (PSNR). PSNR е стандартно използвана метрика за измерване на качеството на изображението и ни дава информация за колко силно шумът влияе на качеството на изображението.

Изведената стойност на коефициента на еднаквост (PSNR) ни казва колко близки са двата файлове до един друг, като по-голяма стойност означава по-голяма сходство между двата файлове.

MSE е акроним за "Mean Squared Error" (средно квадратична грешка). Това е метрика, която се използва за оценка на разликата между два набора от данни, като често се използва за сравнение на оригинални изображения с техните реконструирани версии.

За изображения, MSE може да се изчисли като средноквадратичното отклонение на пикселите на двата набора от данни. Тоест, MSE измерва средният квадрат на разликата между оригиналните и реконструирани пиксели.

MSE може да бъде полезно за сравняване на различни методи за обработка на изображения и оценка на тяхната точност. Той също може да се използва за измерване на точността на прогнози, като например при машинно самообучение, където това е честа метрика за оценка на моделите.

PSNR е съкращение от "Peak Signal-to-Noise Ratio". Това е метрика за измерване на качеството на изображения и звукови данни, която се използва за оценка на загубата на качество при компресията на данни. PSNR се измерва като отношение между максималната възможна мощност на сигнала и средноквадратичното отклонение (MSE) на грешката между оригиналния и компресирания сигнал.

За изображения, PSNR се измерва в dB и по-високата стойност на PSNR означава по-добро качество на изображението. PSNR се използва като метрика за сравнение на качеството на различни методи за компресия на изображения и за определяне на оптималните настройки за компресия.

```python
def calculate_psnr(avg_img, image):
    mse = np.mean((avg_img - image) ** 2)
    if mse == 0:
        psnr = 100
    else:
        max_pixel = 255
        psnr = 20 * np.log10(max_pixel / np.sqrt(mse))
    return psnr
```

```python
psnr_value = calculate_psnr(avg_img, image)
print(f"PSNR: {psnr_value:.2f}")
```

```python
array_of_test_images = np.array([test_01_js, test_02_js, test_03_js, test_04_js, test_05_ws, test_06_kr, test_07_jc, test_08_ip])
```

```python
results = []
for img in array_of_test_images:
    psnr_value = calculate_psnr(avg_img, img)
    results.append({"image": img, "psnr": psnr_value})
```

```python
# print(results)
```

```python
# Създаване на списъци с имената на масивите и техните PSNR стойности
array_names = ["test_01_js", "test_02_js", "test_03_js", "test_04_js", "test_05_ws", "test_06_kr", "test_07_jc", "test_08_ip"]
psnr_values = [result["psnr"] for result in results]

# Построяване на бар диаграма
plt.bar(array_names, psnr_values)

# Задаване на заглавие на диаграмата и маркери на x-ос
plt.title("PSNR values for test images")
plt.xticks(rotation=45)

# Добавяне на стойностите на стълбовете
for i, v in enumerate(psnr_values):
    plt.text(i, v+0.5, str(round(v,2)), horizontalalignment='center', fontweight='bold')

# Показване на диаграмата
plt.show()
```

```python
# Изчисляване на евклидовото разстояние между две изображения
dist = np.linalg.norm(avg_img.flatten() - test_01_js.flatten())

# Отпечатване на резултата
print("Евклидово разстояние : ", dist)
```

$$MSE = \frac{1}{n} \sum_{i=1}^{n}(avgimg_i - image_i)^2$$
Проверяваме дали MSE не е 0, защото в подобен случай ще получим грешка при деление

$$\mathrm{PSNR} = 20 \log_{10} \left( \frac{\mathrm{maxpixel}}{\sqrt{\mathrm{MSE}}} \right)$$

АКО MSE е равно на 0, това означава, че изображенията са напълно еднакви. Тогава няма нужда да сравняваме със PSNR

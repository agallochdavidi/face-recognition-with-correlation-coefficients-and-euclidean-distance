```python
# Създаване на списък с имена на файловете в папката images
files = os.listdir('images-nobkg')

# Зареждане на първото изображение и определяне на размера на матрицата за съхранение на сумата на изображенията
img = Image.open('images-nobkg/'+files[0]).convert('L')
img_arr = np.array(img, dtype=float)
img_sum = np.zeros_like(img_arr, dtype=float)

# Добавяне на всички изображения към сумата и превръщането им в NumPy масиви
for file in files[1:]:
    img = Image.open('images-nobkg/'+file).convert('L')
    img_arr = np.array(img, dtype=float)
    img_sum += img_arr

# Изчисляване на усредненото изображение
avg_img = img_sum / len(files)

# Прескалиране на изображението в диапазона [0, 255] и преобразуване в NumPy масив
avg_img = np.uint8(avg_img / np.max(avg_img) * 255)

# Плотване на усредненото изображение
plt.imshow(avg_img, cmap='gray')
plt.show()
```

```python
avg_img
print(type(avg_img))
print(avg_img.shape)
print(js1.shape)
print(image)
```

```python
def calculate_psnr(avg_img, image):
    mse = np.mean((avg_img - image) ** 2)
    if mse == 0:
        psnr = 100
    else:
        max_pixel = 255
        psnr = 20 * np.log10(max_pixel / np.sqrt(mse))
    return psnr
```

```python
psnr_value = calculate_psnr(avg_img, image)
print(f"PSNR: {psnr_value:.2f}")
```

```python
array_of_test_images = np.array([test_01_js, test_02_js, test_03_js, test_04_js, test_05_ws, test_06_kr, test_07_jc, test_08_ip])
```

```python
results = []
for img in array_of_test_images:
    psnr_value = calculate_psnr(avg_img, img)
    results.append({"image": img, "psnr": psnr_value})
```

```python
# print(results)
```

```python
# Създаване на списъци с имената на масивите и техните PSNR стойности
array_names = ["test_01_js", "test_02_js", "test_03_js", "test_04_js", "test_05_ws", "test_06_kr", "test_07_jc", "test_08_ip"]
psnr_values = [result["psnr"] for result in results]

# Построяване на бар диаграма
plt.bar(array_names, psnr_values)

# Задаване на заглавие на диаграмата и маркери на x-ос
plt.title("PSNR values for test images")
plt.xticks(rotation=45)

# Добавяне на стойностите на стълбовете
for i, v in enumerate(psnr_values):
    plt.text(i, v+0.5, str(round(v,2)), horizontalalignment='center', fontweight='bold')

plt.ylim(28, 32) #Ограничаваме максимума и минимума на графиката
# Показване на диаграмата
plt.show()
```

# Определено този резултат не ме устройва :)

Когато евклидовото растояние, между две изображения е 0, това значи, че са напълно еднакви. Максималната стойност за него е в нашия случай е

$\sqrt{(255^2 \cdot 200^2 \cdot 2)} \approx 18150.7$

$$MSE = \frac{1}{n} \sum_{i=1}^{n}(avgimg_i - image_i)^2$$
Проверяваме дали MSE не е 0, защото в подобен случай ще получим грешка при деление

$$\mathrm{PSNR} = 20 \log_{10} \left( \frac{\mathrm{maxpixel}}{\sqrt{\mathrm{MSE}}} \right)$$

АКО MSE е равно на 0, това означава, че изображенията са напълно еднакви. Тогава няма нужда да сравняваме със PSNR

SSIM (Structural Similarity Index) да видя и какво мамка му е това

MSE може да е от 0 до 255. 0 означава, че са абсолютно еднакви.

```python
# Изчисляване на евклидовото разстояние между две изображения
dist_1 = np.linalg.norm(avg_img.flatten() - test_01_js.flatten())
dist_2 = np.linalg.norm(avg_img.flatten() - test_02_js.flatten())
dist_3 = np.linalg.norm(avg_img.flatten() - test_03_js.flatten())
dist_4 = np.linalg.norm(avg_img.flatten() - test_04_js.flatten())
dist_5 = np.linalg.norm(avg_img.flatten() - test_05_ws.flatten())
dist_6 = np.linalg.norm(avg_img.flatten() - test_06_kr.flatten())
dist_7 = np.linalg.norm(avg_img.flatten() - test_07_jc.flatten())
dist_8 = np.linalg.norm(avg_img.flatten() - test_08_ip.flatten())


# Отпечатване на резултата
print("Евклидово разстояние : ", dist_1)
print("Евклидово разстояние : ", dist_2)
print("Евклидово разстояние : ", dist_3)
print("Евклидово разстояние : ", dist_4)
print("Евклидово разстояние : ", dist_5)
print("Евклидово разстояние : ", dist_6)
print("Евклидово разстояние : ", dist_7)
print("Евклидово разстояние : ", dist_8)
```

```python
distances = [dist_1, dist_2, dist_3, dist_4, dist_5, dist_6, dist_7, dist_8]
labels = ['Distance 1', 'Distance 2', 'Distance 3', 'Distance 4', 'Distance 5', 'Distance 6', 'Distance 7', 'Distance 8']

plt.bar(labels, distances)
plt.xlabel('Distances')
plt.ylabel('Euclidean Distance')
plt.title('Euclidean Distances')
plt.xticks(rotation=45)

plt.show()
```

```python
#Създаваме речник с коефициентите на корелация
correlations = {
    'test_01_js': np.corrcoef(avg_img.flatten(), test_01_js.flatten())[0][1],
    'test_02_js': np.corrcoef(avg_img.flatten(), test_02_js.flatten())[0][1],
    'test_03_js': np.corrcoef(avg_img.flatten(), test_03_js.flatten())[0][1],
    'test_04_js': np.corrcoef(avg_img.flatten(), test_04_js.flatten())[0][1],
    'test_05_ws': np.corrcoef(avg_img.flatten(), test_05_ws.flatten())[0][1],
    'test_06_kr': np.corrcoef(avg_img.flatten(), test_06_kr.flatten())[0][1],
    'test_07_jc': np.corrcoef(avg_img.flatten(), test_07_jc.flatten())[0][1],
    'test_08_ip': np.corrcoef(avg_img.flatten(), test_08_ip.flatten())[0][1]
}

print("Correlation coefficients:")
for test, correlation in correlations.items():
    print(f"{test}: {correlation:.2f}")
```
